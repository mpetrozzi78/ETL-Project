```python
import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
import time
from splinter import Browser
```

```python
#https://splinter.readthedocs.io/en/latest/drivers/chrome.html
!which chromedriver
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
URL = "https://www.indeed.com/jobs?q=data+analyst&l=San+Francisco%2C+CA"
#conducting a request of the stated URL above:
page = requests.get(URL)
page
```

```python
#specifying a desired format of page using the html parser - this allows python to read the various components of the page, rather than treating it as one long string.
soup = BeautifulSoup(page.text,"html.parser")
#printing soup in a more structured tree format that makes for easier reading
print(soup.prettify())
```

```python
# Grab the job posting we are interested in 
job_vacancy = soup.find_all(name="div", attrs={"class":"jobsearch-SerpJobCard"})
print(job_vacancy)
```

```python
data = []
for row in job_vacancy:
    #Error handling
    try:
        print("---------")
        # Scrape the job-title
        job_title = row.find("div", class_= "title")
        if job_title:  
            job_title = job_title.text.strip()
        else:
            job_title = "No job_title"
        
        # Scrape the company
        company_name = row.find("span", class_= "company")
        if company_name:  
            company_name = company_name.text.strip()
        else:
            company_name = "No Company Name"
    
        print(company_name)
        # Scrape the location
        location = row.find("div", class_="location")
        if location:  
            location = location.text.strip()
        else:
            location = "No Location"
        print(location)
    
        # Scrape the salary
        salary = row.find("span", class_="no-wrap")
        if salary:  
            salary = salary.text.strip()
        else:
            salary = "No Salary"
        print(salary)
    
        # Scrape the summary
        summary = row.find("div", class_="summary")
        if (summary):
            summary = summary.text.strip()
        else:
            summary = "No Summary"
        print(summary)
    
        # Scrape the ratings
        rating = row.find("span", class_="ratingsContent")
        if rating:
            rating = rating.text.strip()
        else:
            rating = "No rating"
        print(rating)
    
        # Check if the data is available
        if (job_title and location and salary and summary and rating):
        
            job_dic = {
                "title": job_title,
                "company": company_name,
                "location": location,
                "salary": salary,
                "summary": summary,
                "rating": rating
        
            }
#             print(job_dic)
        data.append(job_dic)
        
#         "collection.insert_one(job_dic)"
    except Exception as e:
        print(e)
        
print(data)
df = pd.DataFrame(data)
df.head()
```

```python
max_results_per_city = 100
city_list = ["New+York","Chicago","San+Francisco", "Austin", "Seattle", "Los+Angeles", "Philadelphia", "Atlanta",
            "Dallas", "Pittsburgh", "Portland", "Phoenix", "Denver", "Houston", "Miami", "Raleigh", "Washington+DC", "Boulder"]
columns = ["company", "location", "rating", "salary", "summary", "title"]
sample_df = pd.DataFrame(columns = columns)
```

```python
# Loop to visit each page in a specific order to extract the data we need
all_data = []
for city in city_list:
    for start in range(0, max_results_per_city, 10):
        url = "https://www.indeed.com/jobs?q=data+analyst&l=" + str(city) + "&start=" + str(start)
        print(url)
        page = requests.get(url)
        time.sleep(1)  #ensuring at least 1 second between page grabs
        soup = BeautifulSoup(page.text,"html.parser")
        job_vacancy = soup.find_all(name="div", attrs={"class":"jobsearch-SerpJobCard"})
        for row in job_vacancy:
            #Error handling
            try:
                print("---------")
                # Scrape the job-title
                job_title = row.find("div", class_= "title")
                if job_title:  
                    job_title = job_title.text.strip()
                else:
                    job_title = "No job_title"
        
                # Scrape the company
                company_name = row.find("span", class_= "company")
                if company_name:  
                    company_name = company_name.text.strip()
                else:
                    company_name = "No Company Name"
    
                print(company_name)
                # Scrape the location
                location = row.find("div", class_="location")
                if location:  
                    location = location.text.strip()
                else:
                    location = "No Location"
                print(location)
    
                # Scrape the salary
                salary = row.find("span", class_="no-wrap")
                if salary:  
                    salary = salary.text.strip()
                else:
                    salary = "No Salary"
                print(salary)
    
                # Scrape the summary
                summary = row.find("div", class_="summary")
                if (summary):
                    summary = summary.text.strip()
                else:
                    summary = "No Summary"
                print(summary)
    
                # Scrape the ratings
                rating = row.find("span", class_="ratingsContent")
                if rating:
                    rating = rating.text.strip()
                else:
                    rating = "No rating"
                print(rating)
    
                # Check if the data is available
                if (job_title and location and salary and summary and rating):
        
                    job_dic = {
                        "title": job_title,
                        "company": company_name,
                        "location": location,
                        "salary": salary,
                        "summary": summary,
                        "rating": rating
        
                    }
                all_data.append(job_dic)
#         "collection.insert_one(job_dic)"
            except Exception as e:
                print(e)
                
                browser.click_link_by_partial_text("next")
            
        
print(all_data)
all_df = pd.DataFrame(all_data)
all_df.head()
```
