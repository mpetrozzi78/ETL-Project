````python
import pandas as pd
from sqlalchemy import create_engine
````

# Extract


````python
# Read csv file of scraped data analyst jobs from Indeed
csv1_file = "scraped_jobs.csv"
scraped_df = pd.read_csv(csv1_file)
scraped_df.head()
````

````python
# Read csv file of all data
csv2_file = "Resources/alldata.csv"
all_data_df = pd.read_csv(csv2_file)
all_data_df.head()
````

````python
# Read csv file of cost of living 2018
csv3_file = "Resources/cost-of-living-2018.csv"
cost_living_df = pd.read_csv(csv3_file)
cost_living_df.head()
````

# CLEANUP


````python
# Clean up scraped_df dataframe 2019

# dropping null value columns to avoid errors 
scraped_df.dropna(inplace = True)
scraped_df.head()
````

````python
# new data frame with split value columns 
new = scraped_df["location"].str.split(",", n = 1, expand = True)
# making separate city column from new data frame 
scraped_df["city"]= new[0] 
  
# making separate state column from new data frame 
scraped_df["state_zip"]= new[1] 

# Dropping old location column 
scraped_df.drop(columns =["location"], inplace = True) 
  
# df display 
scraped_df
````

````python
# Dropping duplicates
scraped_df.drop_duplicates("summary", inplace=True)
# Set index
scraped_df.set_index("city", inplace=True)
````

````python
scraped_df.head()
````

````python
# Cleanup of cost_living_df 2018

# new data frame with split value columns 
new = cost_living_df["City"].str.split(",", n = 1, expand = True)

# making separate city column from new data frame 
cost_living_df["city"]= new[0] 
  
# making separate state column from new data frame 
cost_living_df["country"]= new[1] 

# Dropping old location column 
cost_living_df.drop(columns =["City"], inplace = True) 
  
# df display 
cost_living_df.head()
````

````python
cost_living_df = cost_living_df[['city', 'country', 'Cost of Living Index', 'Rent Index', 'Groceries Index',
                                'Restaurant Price Index', 'Local Purchasing Power Index']].copy()
cost_living_df.head()
````

````python
# Dropping duplicates
cost_living_df.drop_duplicates("city", inplace=True)
# Set index
cost_living_df.set_index("city", inplace=True)
cost_living_df.head()
````

````python
# rds_connection_string = "<insert user name>:<insert password>@localhost:5432/ETL_db"
rds_connection_string = "postgres:postgres@localhost:5432/ETL_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
````

````python
engine.table_names()
````

````python
scraped_df.to_sql(name='city', con=engine, if_exists='replace', index=False)
````
